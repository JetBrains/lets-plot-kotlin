# Tooltip Customization API: `disableSplitting()`

New function `disableSplitting()` allows to hide side tooltips.

By default, this function moves all side tooltips to the general tooltip. If the content of a general tooltip is specified with the `line()` functions, the general tooltip will get the given lines, and the side tooltips will be hidden.

```kotlin
%useLatestDescriptors
%use lets-plot
%use dataframe
```

```kotlin
LetsPlot.getInfo()
```

```kotlin
var mpgDf = DataFrame.readCSV("https://raw.githubusercontent.com/JetBrains/lets-plot-kotlin/master/docs/examples/data/mpg.csv")
val mpgData = mpgDf.toMap()
mpgDf.head()
```

#### 1. Default Tooltips

```kotlin
letsPlot(mpgData) {x="class"; y="hwy"} + geomBoxplot()
```

#### `disableSplitting()`

The function moves all side tooltips to the general tooltip.

```kotlin
letsPlot(mpgData) {x="class"; y="hwy"} + geomBoxplot(tooltips = layerTooltips().disableSplitting())
```

#### 3. `disableSplitting()` with Specified Tooltip Lines

All side tooltips are hidden, the general tooltip gets the specified lines.

```kotlin
letsPlot(mpgData) {x="class"; y="hwy"} + geomBoxplot(tooltips = layerTooltips()
                                                                .line("@|^middle")
                                                                .disableSplitting())
```
