# Auto-detected `orientation="y"`


```kotlin
%useLatestDescriptors
%use lets-plot
%use dataframe
```

```kotlin
LetsPlot.getInfo()
```

```kotlin
val df = DataFrame.readCSV("https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/gapminder.csv")
val albaniaDf = df.filter { it["country"] == "Albania" }

df.head(3)
```

```kotlin
val data = df.toMap()
val albaniaData = albaniaDf.toMap()
```

#### 1. How Does Auto-detection of `orientation="y"` Work

```kotlin
gggrid(listOf(
    letsPlot(data) { x = "continent"; y = "pop"; fill = "continent" } + geomBar() + ggtitle("Default"),
    letsPlot(data) { x = "pop"; y = "continent"; fill = "continent" } + geomBar() + ggtitle("Automatic setting orientation=\"y\"")
), ncol = 2)
```

#### 2. Marking by `asDiscrete()`

If you're using continuous data for the Y-axis, the plot may be incorrect. Try marking the Y-axis data with the `asDiscrete()` function, this will rotate your geom.

```kotlin
gggrid(listOf(
    letsPlot(albaniaData) + geomBar(stat = Stat.sum(), size = 0) { x = "lifeExp"; y = "year" },
    letsPlot(albaniaData) + geomBar(stat = Stat.sum(), size = 0) { x = "lifeExp"; y = asDiscrete("year") }
), ncol = 2)
```

#### 3. Other "Oriented" Geometries are Also Supported

```kotlin
gggrid(listOf(
    letsPlot(data) { x = "continent"; y = "lifeExp"; fill = "continent" } + geomBoxplot(),
    letsPlot(data) { x = "lifeExp"; y = "continent"; fill = "continent" } + geomBoxplot(),
    letsPlot(data) { x = "continent"; y = "lifeExp"; fill = "continent" } + geomViolin(),
    letsPlot(data) { x = "lifeExp"; y = "continent"; fill = "continent" } + geomViolin(),
    letsPlot(albaniaData) { x = asDiscrete("year"); y = "lifeExp" } + geomLollipop(stat = Stat.sum(), size = 2),
    letsPlot(albaniaData) { x = "lifeExp"; y = asDiscrete("year") } + geomLollipop(stat = Stat.sum(), size = 2),
), ncol = 2)
```

#### 4. `stat = Stat.boxplot()`, `stat = Stat.boxplotOutlier()`, `statSummary()`

```kotlin
gggrid(listOf(
    letsPlot(data) { x = "continent"; y = "lifeExp" } + geomPointRange(stat = Stat.boxplot()),
    letsPlot(data) { x = "lifeExp"; y = "continent" } + geomPointRange(stat = Stat.boxplot()),
    letsPlot(data) { x = "continent"; y = "lifeExp" } + geomPointRange(stat = Stat.boxplot()),
    letsPlot(data) { x = "lifeExp"; y = "continent" } + geomPointRange(stat = Stat.boxplotOutlier()),
    letsPlot(data) { x = "continent"; y = "lifeExp" } + statSummary(),
    letsPlot(data) { x = "lifeExp"; y = "continent" } + statSummary()
), ncol = 2)
```
