```kotlin
val random = java.util.Random(42)
val bigData = mapOf(
    "x" to List(75_000) { random.nextGaussian() },
    "y" to List(75_000) { random.nextGaussian() }
)

letsPlot(bigData) +
    geomPoint(sampling = samplingRandom(500, seed = 42) + samplingSystematic(100)) {
        x = "x"
        y = "y"
    }
```
