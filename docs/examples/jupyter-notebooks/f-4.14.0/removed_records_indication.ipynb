# Indication of Removed Records

Lets-Plot-Kotlin provides informative messages when records are excluded during plot preparation.

These notifications specify number of dropped records and the reason for their removal, to
ensure that the final visualization is interpreted correctly.

Records can be removed due to:

**Sampling** - reducing an oversized dataset for performance.

**Statistics & Geometries** - filtering out non-finite values (NaN, Inf) or out-of-bounds data.

```kotlin
%useLatestDescriptors
%use lets-plot
```

```kotlin
LetsPlot.getInfo()
```

## Sampling

Sampling helps dealing with excessively large datasets by reducing the number of data points rendered.

Learn more: [Sampling in Lets-Plot-Kotlin](https://lets-plot.org/kotlin/sampling.html).

```kotlin
val random = java.util.Random(42)
val bigData = mapOf(
    "x" to List(75_000) { random.nextGaussian() },
    "y" to List(75_000) { random.nextGaussian() }
)

letsPlot(bigData) +
    geomPoint(sampling = samplingRandom(500, seed = 42) + samplingSystematic(100)) {
        x = "x"
        y = "y"
    }
```

## Statistics

Stats drop non-finite values before computing transformations. The resulting message identifies the stat responsible for the removal.

```kotlin
val ridgesData = mapOf(
    "x" to listOf(1.0, 2.0, Double.NaN, 4.0, 1.0, 2.0, 3.0, Double.NaN),
    "y" to listOf(0, 0, 0, 0, 1, 1, 1, 1)
)

letsPlot(ridgesData) + geomAreaRidges {
    x = "x"
    y = "y"
}
```

## Geometries

Geoms remove records that contain missing values or fall outside the scale limits.


```kotlin
val data = mapOf(
    "x" to listOf(1.0, 2.0, Double.NaN, 4.0, 5.0, Double.NaN, 7.0, 8.0),
    "y" to listOf(2.0, Double.NaN, 3.0, 4.0, 5.0, Double.NaN, 7.0, 8.0)
)

letsPlot(data) + geomPoint {
    x = "x"
    y = "y"
}
```

## Suppressing Messages

While notifications are shown by default, you can control this behavior:

**Per Layer:** Via `naRm` parameter:

- `naRm = false` (default) - records are removed and messages are shown;
- `naRm = true` - records are removed silently.


**Per Plot or Globally:** Use `theme(plotMessage = elementBlank())` to silence all notifications.

Note: you can suppress messages globally using `LetsPlot.theme = theme(plotMessage = elementBlank())`.

```kotlin
// Suppress messages from a layer
letsPlot(data) + geomPoint(naRm = true) {
    x = "x"
    y = "y"
}
```

```kotlin
// Suppress messages from all layers
letsPlot(ridgesData) +
    geomAreaRidges {
        x = "x"
        y = "y"
    } +
    theme(plotMessage = elementBlank())
```
