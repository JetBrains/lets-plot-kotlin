# Legend Key Parameters in `theme()`
New parameters in `theme()` to customize the legend key:

- `legendKey` - background underneath legend keys, set with `elementRect()`
- `legendKeySize` - size of legend keys
- `legendKeyWidth` - key background width
- `legendKeyHeight` - key background height
- `legendKeySpacing` - spacing between legend keys
- `legendKeySpacingX` - spacing in the horizontal direction
- `legendKeySpacingY` - spacing in the vertical direction

```kotlin
%useLatestDescriptors
%use lets-plot

%use dataframe
```

```kotlin
LetsPlot.getInfo()
```

```kotlin
val mpg = DataFrame.readCSV("https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/mpg.csv")
mpg.head(3)
```

```kotlin
val p = letsPlot(mpg.toMap()) { x = "hwy" } +
    geomDotplot(color="pen") { fill = "class" } +
    theme(legendBackground = elementRect(size = 0.5))
p
```

Add background underneath legend keys:

```kotlin
val p1 = p + theme(legendKey = elementRect(fill = "#efedf5", color = "#756bb1"))
p1
```

Change size of legend keys:

```kotlin
p1 + theme(legendKeySize = 30)
```

```kotlin
val p2 = p1 + theme(legendKeyWidth = 80, legendKeyHeight = 20)
p2
```

Add spacing between legend keys:

```kotlin
p2 + theme(legendKeySpacingY = 10)
```
