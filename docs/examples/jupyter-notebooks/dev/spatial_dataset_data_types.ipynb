## Attribute Values

```kotlin
import java.awt.Color
import java.time.Duration
import java.time.LocalDate
import java.time.LocalDateTime
import java.time.LocalTime

val rows = (1..7).toList()
val attributeTypes = SpatialDataset.withGEOJSON(
    data = mapOf(
        "string" to rows.map { "label-$it" },
        "escaped" to listOf("<tag>", "A & B", "\"quoted\"", "'single'", "plain", "dash - value", "tail"),
        "int" to IntArray(rows.size) { it + 1 },
        "long" to LongArray(rows.size) { 1_000_000_000L + it },
        "double" to rows.map { it + 0.125 },
        "float" to FloatArray(rows.size) { (it + 1) / 10f },
        "boolean" to rows.map { it % 2 == 0 },
        "char" to rows.map { ('A'.code + it - 1).toChar() },
        "enum" to rows.map { Thread.State.values()[(it - 1) % Thread.State.values().size] },
        "color" to rows.map { Color(20 * it, 12 * it, 30 * it) },
        "date" to rows.map { LocalDate.of(2026, 1, it) },
        "time" to rows.map { LocalTime.of(it + 8, 30) },
        "date_time" to rows.map { LocalDateTime.of(2026, 1, it, 9, 15) },
        "duration" to rows.map { Duration.ofMinutes((it * 15).toLong()) },
        "nullable" to listOf("text", null, 42, null, true, 3.14, "end"),
        "list" to rows.map { listOf(it, it * 2) },
        "map" to rows.map { mapOf("id" to it, "even" to (it % 2 == 0)) },
        "long_text" to rows.map { "row-$it-" + "x".repeat(60) }
    ),
    geometry = rows.map {
        "{\"type\":\"Point\",\"coordinates\":[${-74.0 + it * 0.01},${40.7 + it * 0.01}]}"
    }
)

attributeTypes
```

## GeoJSON Geometry Types

```kotlin
val geoJsonTypes = SpatialDataset.withGEOJSON(
    data = mapOf(
        "type" to listOf("Point", "MultiPoint", "LineString", "MultiLineString", "Polygon", "MultiPolygon", "GeometryCollection")
    ),
    geometry = listOf(
        "{\"type\":\"Point\",\"coordinates\":[-73.9857,40.7484]}",
        "{\"type\":\"MultiPoint\",\"coordinates\":[[-73.9857,40.7484],[-73.9654,40.7829]]}",
        "{\"type\":\"LineString\",\"coordinates\":[[-74.0122,40.7061],[-73.9969,40.7223],[-73.9822,40.7532]]}",
        "{\"type\":\"MultiLineString\",\"coordinates\":[[[-74.0122,40.7061],[-73.9969,40.7223]],[[-73.9822,40.7532],[-73.9654,40.7829]]]}",
        "{\"type\":\"Polygon\",\"coordinates\":[[[-74.0122,40.7061],[-73.9969,40.7223],[-73.9822,40.7532],[-74.0122,40.7061]]]}",
        "{\"type\":\"MultiPolygon\",\"coordinates\":[[[[-74.0122,40.7061],[-73.9969,40.7223],[-73.9822,40.7532],[-74.0122,40.7061]]],[[[-73.9857,40.7484],[-73.9654,40.7829],[-74.003,40.7506],[-73.9857,40.7484]]]]}",
        "{\"type\":\"GeometryCollection\",\"geometries\":[{\"type\":\"Point\",\"coordinates\":[-73.9857,40.7484]},{\"type\":\"LineString\",\"coordinates\":[[-74.0122,40.7061],[-73.9969,40.7223]]}]}"
    )
)

geoJsonTypes
```

## WKT Geometry Format

```kotlin
val wktFormat = SpatialDataset.withWKT(
    data = mapOf(
        "id" to rows,
        "label" to rows.map { "wkt-$it" }
    ),
    geometry = rows.map { "POINT (${it}.0 ${it + 1}.0)" }
)

wktFormat
```

## WKB Geometry Format

```kotlin
val wkbFormat = SpatialDataset.withWKB(
    data = mapOf(
        "id" to rows,
        "label" to rows.map { "wkb-$it" }
    ),
    geometry = rows.map { "0101000000000000000000F03F0000000000000040" }
)

wkbFormat
```
