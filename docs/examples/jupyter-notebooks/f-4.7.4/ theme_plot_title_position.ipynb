# Plot Title/Subtitle and Caption Positions

The `plotTitlePosition` and `plotCaptionPosition` parameters set the alignment of the plot title/subtitle and caption.

A value of `"panel"` (the default) means that titles/caption are aligned with the plot panel. A value of `"plot"` means that titles/caption are aligned to the entire plot, excluding margins.


```kotlin
%useLatestDescriptors
%use lets-plot

%use dataframe
```

```kotlin
LetsPlot.getInfo()
```

```kotlin
val mpg = DataFrame.readCSV("https://raw.githubusercontent.com/JetBrains/lets-plot-docs/master/data/mpg.csv")
```

```kotlin
val p = letsPlot(mpg.toMap()) +
    geomPoint() { x = "displ"; y = "cty"; color = "drv" } +
    labs(title = "Title of the plot",
         subtitle = "Subtitle of the plot",
         caption = "Caption of the plot")
```

By default titles and caption are aligned with the plot panel:

```kotlin
p
```

The position can be set relative to the entire plot using the `plotTitlePosition` and `plotCaptionPosition` parameters of the theme function:

```kotlin
p + theme(plotTitlePosition = "plot", plotCaptionPosition = "plot")
```
