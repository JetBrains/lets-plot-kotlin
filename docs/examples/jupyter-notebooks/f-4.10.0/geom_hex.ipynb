# Hexagonal Heatmap

```kotlin
%useLatestDescriptors
%use dataframe
%use lets-plot
```

```kotlin
LetsPlot.getInfo()
```

```kotlin
val df = DataFrame.readCSV("https://github.com/JetBrains/lets-plot-docs/raw/refs/heads/master/data/diamonds.csv")
val dataMap = df.toMap()
df.head()
```

## Default View

```kotlin
val p = letsPlot(dataMap) { x = "carat"; y = "price" } + scaleFillViridis(trans = "symlog")
```

```kotlin
p + geomHex()
```

## Parameters

```kotlin
p + geomHex(bins = Pair(5, 10)) + ggtitle("bins = Pair(5, 10)")
```

```kotlin
p + geomHex(binWidth = Pair(1, 5000)) + ggtitle("binWidth = Pair(1, 5000)")
```

```kotlin
p + geomHex(binWidth = Pair(1, 5000), drop = false) + ggtitle("drop = false")
```

## `Stat.binhex()`

```kotlin
letsPlot(dataMap) { x = "carat"; y = "price" } +
    geomPoint(stat = Stat.binhex(binWidth = Pair(1, 5000), drop = false),
              size = .75, sizeUnit = "x") { color = "..count.." } +
    coordCartesian(xlim = Pair(-1, 7), ylim = Pair(-2_500, 25_000))
```
