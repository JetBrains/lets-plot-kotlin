# `geomSpoke()`

```kotlin
%useLatestDescriptors
%use lets-plot
```

```kotlin
LetsPlot.getInfo()
```

#### 1. Basic Usage

```kotlin
fun getData(n: Int): Map<String, Any> {
    val rand = java.util.Random()
    val x = ArrayList<Double>()
    val y = ArrayList<Double>()
    val z = ArrayList<Double>()
    val angle = ArrayList<Double>()
    (-n..n).toList().forEach { row ->
        (-n..n).toList().forEach { col ->
            x.add(col.toDouble())
            y.add(row.toDouble())
            z.add(rand.nextGaussian())
            angle.add(PI/2 - col*row*PI/4)
        }
    }
    return mapOf(
        "x" to x, 
        "y" to y, 
        "z" to z, 
        "angle" to angle
    )
}
```

```kotlin
val data = getData(n = 10)
```

```kotlin
val p = letsPlot(data) { x = "x"; y = "y"  } + coordFixed()
```

```kotlin
p + geomBin2D(stat = Stat.identity) { fill = "z" }
```

```kotlin
p + geomPoint(size = 1.5) + geomSpoke(radius=0.5) { angle = "angle"; color = "z" }
```

#### 2. Parameter `pivot`

```kotlin
fun getPlot(pivot: String): org.jetbrains.letsPlot.intern.Plot {
    val pivotData = getData(1)
    val title = "pivot=${pivot}" + (" (default)".takeIf { pivot == "tail" } ?: "")
    return ggplot(pivotData) { x = "x"; y = "y" } +
        geomPoint() +
        geomSpoke(radius = 0.5, pivot = pivot) { angle = "angle" } +
        coordFixed() +
        ggtitle(title)
}
```

```kotlin
gggrid(
    listOf(getPlot("tail"), getPlot("mid"), getPlot("tip")),
    ncol=3
)
```
